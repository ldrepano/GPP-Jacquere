### Picking analysis: Current CRISPick
To show current performance of CRISPick, ran with default settings (i.e. off-target to on-target weights of 1:1), Ensembl annotations, RS3 Seq+Target Chen, library mode with all sequences in rs3val_ess_noness_lfcs, list of all genes targeted, quota=1 and "show unpicked sequences" 

```python
current_crispick= pd.read_csv("../Data/rs3val_ess_and_noness_defaults_01_22_2024-sgrna-designs.txt",sep='\t',header=0)
#orphaned preselects don't have any info. honestly not sure what these rows are about
current_crispick=current_crispick[current_crispick["Picking Notes"]!="Orphaned Preselect"].reset_index(drop=True).copy()
#identify screening behavior of each sgRNA and whether it is essential or nonessential targeting
rs3val["Gene and Sequence"]=rs3val["Target Gene Symbol"]+rs3val["sgRNA Sequence"]
current_crispick["Gene and Sequence"]=current_crispick["Target Gene Symbol"]+current_crispick["sgRNA Sequence"]
current_crispick=current_crispick.merge(rs3val[["Gene and Sequence","Category","z_scored_sgRNA_lfc"]],on="Gene and Sequence")
current_crispick
```

```python
#remove non-numeric pick orders 
current_crispick=current_crispick[pd.to_numeric(current_crispick['Pick Order'], errors='coerce').notnull()].reset_index(drop=True).copy()
#adjust on-target rank after merge
current_crispick["On-Target Rank"]=current_crispick.groupby('Target Gene Symbol')['On-Target Rank'].rank("min")

current_crispick_top5=current_crispick[current_crispick["Pick Order"]<=5].reset_index(drop=True).copy()

#top 5 picks for essentials
current_crispick_essential_top5= current_crispick_top5[current_crispick_top5["Category"]=="Essential"].reset_index(drop=True).copy()
#top 5 picks for nonessentials
current_crispick_nonessential_top5=current_crispick_top5[current_crispick_top5["Category"]=="Non-essential"].reset_index(drop=True).copy()
```

```python
#update core dataframe to have on-target rank of guides targeting each gene
rs3val=current_crispick[["Gene and Sequence","Target Gene Symbol","Category","z_scored_sgRNA_lfc","On-Target Rank"]]
```

### Picking analysis: My proposal 
To show performance of CRISPick with Agg CFD cutoff, ran with off-target rank weight of 0, Ensembl annotations, RS3 Seq+Target Chen, library mode with all sequences in rs3val_ess_noness_lfcs, list of all genes targeted, quota=1 and "show unpicked sequences". sgRNAs should then be filtered by agg CFD 0-1 MM and pick order re-adjusted. 

```python
#loading in essentials,nonessentials, off-target rank weight=0 regular CRISPick run 
no_offtarget_crispick= pd.read_csv("../Data/rs3val_ess_noness_offtargetrankweight0_1_22_2024-sgrna-designs.txt",sep='\t',header=0)

#repeat same cleaning
no_offtarget_crispick=no_offtarget_crispick[no_offtarget_crispick["Picking Notes"]!="Orphaned Preselect"].reset_index(drop=True).copy()
no_offtarget_crispick=no_offtarget_crispick[no_offtarget_crispick["On-Target Ruleset"]=="RS3seq-Chen2013+RS3target"].reset_index(drop=True).copy()
no_offtarget_crispick["Gene and Sequence"]=no_offtarget_crispick["Target Gene Symbol"]+no_offtarget_crispick["sgRNA Sequence"]
#merge w/ screening data
no_offtarget_crispick=no_offtarget_crispick.merge(rs3val[["Gene and Sequence",'Category',"z_scored_sgRNA_lfc"]],on="Gene and Sequence")
#remove non-numeric pick orders 
no_offtarget_crispick=no_offtarget_crispick[pd.to_numeric(no_offtarget_crispick['Pick Order'], errors='coerce').notnull()].reset_index(drop=True).copy()

no_offtarget_crispick
```

Getting aggregate CFD on 0-1MM for essential sgRNAs (already have nonessentials) from verbose CRISPick runs. These jobs were run separately because they were too large to run in 1 go.  

```python
#essentials verbose batch 1
with gzip.open('../Data/rs3validation_essentials_batch1_verbose_1_22_2024-sgrna-designs.offtargetdisco.txt.gz') as f:
    rs3val_ess_verbose_batch1 = pd.read_table(f,index_col=False,header=None,names=["Target","Target Sequence","Context Sequence","Match Tier", "CFD Score","PAM Sequence", "# of mismatches"])
#essentials verbose batch 2
with gzip.open('../Data/rs3val_essentials_batch2_verbose_1_22_2024-sgrna-designs.offtargetdisco.txt.gz') as f:
    rs3val_ess_verbose_batch2 = pd.read_table(f,index_col=False,header=None,names=["Target","Target Sequence","Context Sequence","Match Tier", "CFD Score","PAM Sequence", "# of mismatches"])
#join both essential batches
rs3val_ess_verbose=pd.concat([rs3val_ess_verbose_batch1,rs3val_ess_verbose_batch2]).reset_index(drop=True)
rs3val_ess_verbose
```

```python
#nonessentials verbose
with gzip.open('../Data/sgrna-designs.targetdisco.txt.gz') as f:
    rs3val_noness_verbose = pd.read_table(f,index_col=False,header=None,names=["Target","Target Sequence","Context Sequence","Match Tier", "CFD Score","PAM Sequence", "# of mismatches"])
#join essentials and nonessentials
rs3val_verbose=pd.concat([rs3val_ess_verbose,rs3val_noness_verbose])
```

```python

#get rid of NA pick orders (i.e. MAX OTS)
rs3val_verbose=rs3val_verbose[pd.to_numeric(rs3val_verbose["# of mismatches"], errors='coerce').notnull()].reset_index(drop=True).copy()
rs3val_verbose["# of mismatches"]=rs3val_verbose["# of mismatches"].astype(np.int64)

#calculate Agg CFD 0-1MM
rs3val_verbose["CFD (0 if >1 mm)"]= np.where(rs3val_verbose["# of mismatches"]<=1, rs3val_verbose["CFD Score"].astype(np.float64),0)
rs3val_agg = (rs3val_verbose.groupby("Target Sequence")
            .agg(cfd_sum_mm0_mm1_sites= ("CFD (0 if >1 mm)","sum"))
            .reset_index())
rs3val_agg
```

```python
#add in agg CFD scores to CRISPick results with 0 off-target rank weight
both_metrics=no_offtarget_crispick.merge(rs3val_agg,left_on="sgRNA Sequence",right_on="Target Sequence")
```

```python
#set Agg CFD Cutoff:
agg_cfd_cutoff= 4.2
#filter out sgRNAs that exceed cutoff
agg_cfd_picks=both_metrics[both_metrics["cfd_sum_mm0_mm1_sites"]<agg_cfd_cutoff].reset_index(drop=True).copy()
#adjust ranks
agg_cfd_picks["Pick Order after Agg CFD Cutoff"]=agg_cfd_picks.groupby('Target Gene Symbol')['Pick Order'].rank("min")

#subset top 5, separate essentials and nonessentials
agg_cfd_top5=agg_cfd_picks[agg_cfd_picks["Pick Order after Agg CFD Cutoff"]<=5].reset_index(drop=True).copy()
#top 5 picks for essentials
agg_cfd_essential_top5= agg_cfd_top5[agg_cfd_top5["Category"]=="Essential"].reset_index(drop=True).copy()
#top 5 picks for nonessentials
agg_cfd_nonessential_top5=agg_cfd_top5[agg_cfd_top5["Category"]=="Non-essential"].reset_index(drop=True).copy()

```

```python
#library level analysis

# % of selected sgRNAs that are promiscuous
perc_prom_aggcfd_relaxed=100*len(agg_cfd_nonessential_top5[agg_cfd_nonessential_top5['z_scored_sgRNA_lfc']<-2])/len(agg_cfd_nonessential_top5)
perc_prom_rank=100*len(current_crispick_nonessential_top5[current_crispick_nonessential_top5['z_scored_sgRNA_lfc']<-2])/len(current_crispick_nonessential_top5)


# % of selected sgRNAs in top5 on-target ranking 
perc_active_aggcfd_relaxed= 100*len(agg_cfd_top5[agg_cfd_top5["On-Target Rank"]<=5])/len(agg_cfd_top5)
perc_active_rank= 100*len(current_crispick_top5[current_crispick_top5["On-Target Rank"]<=5])/len(current_crispick_top5)

```

```python
## repeating with stricter Agg CFD cutoff to show why we don't want this 



#set stricter Agg CFD Cutoff:
agg_cfd_cutoff_strict= 1
#filter out sgRNAs that exceed cutoff
agg_cfd_picks_strict=both_metrics[both_metrics["cfd_sum_mm0_mm1_sites"]<agg_cfd_cutoff_strict].reset_index(drop=True).copy()
#adjust ranks
agg_cfd_picks_strict["Pick Order after Agg CFD Cutoff"]=agg_cfd_picks_strict.groupby('Target Gene Symbol')['Pick Order'].rank("min")

#subset top 5, separate essentials and nonessentials
agg_cfd_top5_strict=agg_cfd_picks_strict[agg_cfd_picks_strict["Pick Order after Agg CFD Cutoff"]<=5].reset_index(drop=True).copy()
#top 5 picks for nonessentials
agg_cfd_nonessential_top5_strict=agg_cfd_top5_strict[agg_cfd_top5_strict["Category"]=="Non-essential"].reset_index(drop=True).copy()







#library level analysis

# % of selected sgRNAs that are promiscuous
perc_prom_aggcfd_strict=100*len(agg_cfd_nonessential_top5_strict[agg_cfd_nonessential_top5_strict['z_scored_sgRNA_lfc']<-2])/len(agg_cfd_nonessential_top5_strict)
perc_prom_rank=100*len(current_crispick_nonessential_top5[current_crispick_nonessential_top5['z_scored_sgRNA_lfc']<-2])/len(current_crispick_nonessential_top5)


# % of selected sgRNAs in top5 on-target ranking 
perc_active_aggcfd_strict= 100*len(agg_cfd_top5_strict[agg_cfd_top5_strict["On-Target Rank"]<=5])/len(agg_cfd_top5_strict)
perc_active_rank=100* len(current_crispick_top5[current_crispick_top5["On-Target Rank"]<=5])/len(current_crispick_top5)


```

### Picking analysis: Guidescan universal cutoff


Guidescan filters out all sgRNAs with:

- any OTS with 0 mismatches to target site

- more than 1 OTS with 1 mismatch to target site 

- CFD sum of OTS with up to 3 mismatches of >=4 (this is equivalent to specificity score 0.2) 

and includes no OTS with an NGG PAM in this search. Mismatch # includes any mismatch in entire 20mer.


```python
guidescan_batch1=pd.read_csv("../Data/rs3val_ess_noness_guidescan_batch1.csv")
guidescan_batch2=pd.read_csv("../Data/rs3val_ess_noness_guidescan_batch2.csv")
guidescan_batch3=pd.read_csv("../Data/rs3val_ess_noness_guidescan_batch3.csv")

guidescan=pd.concat([guidescan_batch1,guidescan_batch2]).reset_index(drop=True)
guidescan=pd.concat([guidescan,guidescan_batch3]).reset_index(drop=True)
guidescan
```

```python
#specificity=0 if it violates the 0 or 1mm rule
guidescan["guidescan_specificity"]=np.where(guidescan["Chromosome"].str.contains("chr"), guidescan["Specificity"],0)
guidescan["sgRNA Sequence"]=guidescan["gRNA"].str[:20]
#add to no off-target rank crispick results 
no_offtarget_crispick_withguidescan=no_offtarget_crispick.merge(guidescan,on="sgRNA Sequence")
both_metrics=both_metrics.merge(guidescan,on="sgRNA Sequence")
```

```python
#set Guidescan Cutoff:
guidescan_cutoff= 0.2
#filter out sgRNAs below cutoff
guidescan_picks=no_offtarget_crispick_withguidescan[no_offtarget_crispick_withguidescan["guidescan_specificity"]>guidescan_cutoff].reset_index(drop=True).copy()
#adjust ranks
guidescan_picks["Pick Order after Guidescan Cutoff"]=guidescan_picks.groupby('Target Gene Symbol')['Pick Order'].rank("min")
#subset top 5, separate essentials and nonessentials
guidescan_top5=guidescan_picks[guidescan_picks["Pick Order after Guidescan Cutoff"]<=5].reset_index(drop=True).copy()
#top 5 picks for essentials
guidescan_essential_top5= guidescan_top5[guidescan_top5["Category"]=="Essential"].reset_index(drop=True).copy()
#top 5 picks for nonessentials
guidescan_nonessential_top5=guidescan_top5[guidescan_top5["Category"]=="Non-essential"].reset_index(drop=True).copy()
```

```python
#fraction of all sgRNAs that guidescan classifies promiscuous
1-(len(guidescan_picks)/len(both_metrics))
```

```python
#2/14 addition: adding guidescan to library metrics plot (same base as agg cfd library, exclude ots with guidescan cutoff)
guidescan_nonessential_top5

perc_prom_guidescan=100*len(guidescan_nonessential_top5[guidescan_nonessential_top5['z_scored_sgRNA_lfc']<-2])/len(guidescan_nonessential_top5)
perc_active_guidescan= 100*len(guidescan_top5[guidescan_top5["On-Target Rank"]<=5])/len(guidescan_top5)
print("guidescan library (nonessentials) are", round(perc_prom_guidescan,3),"% promsicuous, and", round(perc_active_guidescan,3),"% of the whole guidescan library is in the top 5 most active sgRNAs")
```

```python
# 2/14 addition: plotting with continuous aggregate CFD cutoff
perc_prom_rank=100*len(current_crispick_nonessential_top5[current_crispick_nonessential_top5['z_scored_sgRNA_lfc']<-2])/len(current_crispick_nonessential_top5)
perc_active_rank= 100*len(current_crispick_top5[current_crispick_top5["On-Target Rank"]<=5])/len(current_crispick_top5)

thresholds=[i*.1 for i in range(1,50)]

aggcfd_percent_prom_list=[]
aggcfd_percent_active_list=[]
for threshold in thresholds:
    #filter out sgRNAs that exceed cutoff
    agg_cfd_picks_strict=both_metrics[both_metrics["cfd_sum_mm0_mm1_sites"]<threshold].reset_index(drop=True).copy()
    #adjust ranks
    agg_cfd_picks_strict["Pick Order after Agg CFD Cutoff"]=agg_cfd_picks_strict.groupby('Target Gene Symbol')['Pick Order'].rank("min")
    #subset top 5, separate essentials and nonessentials
    agg_cfd_top5_strict=agg_cfd_picks_strict[agg_cfd_picks_strict["Pick Order after Agg CFD Cutoff"]<=5].reset_index(drop=True).copy()
    #top 5 picks for nonessentials
    agg_cfd_nonessential_top5_strict=agg_cfd_top5_strict[agg_cfd_top5_strict["Category"]=="Non-essential"].reset_index(drop=True).copy()
    
    # % of selected sgRNAs that are promiscuous
    aggcfd_percent_prom_list.append(100*len(agg_cfd_nonessential_top5_strict[agg_cfd_nonessential_top5_strict['z_scored_sgRNA_lfc']<-2])/len(agg_cfd_nonessential_top5_strict))
    # % of selected sgRNAs in top5 on-target ranking 
    aggcfd_percent_active_list.append(100*len(agg_cfd_top5_strict[agg_cfd_top5_strict["On-Target Rank"]<=5])/len(agg_cfd_top5_strict))

```

```python
#focusing on 3.7 threshold (just before Agg CFD exceeds promiscuity of rank-based)
#filter out sgRNAs that exceed cutoff
agg_cfd_picks_strict=both_metrics[both_metrics["cfd_sum_mm0_mm1_sites"]<3.7].reset_index(drop=True).copy()
#adjust ranks
agg_cfd_picks_strict["Pick Order after Agg CFD Cutoff"]=agg_cfd_picks_strict.groupby('Target Gene Symbol')['Pick Order'].rank("min")
#subset top 5, separate essentials and nonessentials
agg_cfd_top5_strict=agg_cfd_picks_strict[agg_cfd_picks_strict["Pick Order after Agg CFD Cutoff"]<=5].reset_index(drop=True).copy()
#top 5 picks for nonessentials
agg_cfd_nonessential_top5_strict=agg_cfd_top5_strict[agg_cfd_top5_strict["Category"]=="Non-essential"].reset_index(drop=True).copy()

# % of selected sgRNAs that are promiscuous
aggcfd_percent_prom=100*len(agg_cfd_nonessential_top5_strict[agg_cfd_nonessential_top5_strict['z_scored_sgRNA_lfc']<-2])/len(agg_cfd_nonessential_top5_strict)
# % of selected sgRNAs in top5 on-target ranking 
aggcfd_percent_active=100*len(agg_cfd_top5_strict[agg_cfd_top5_strict["On-Target Rank"]<=5])/len(agg_cfd_top5_strict)

```

Altering to reflect random selection

```python
random.seed(1)
sgRNAs_selected=[]
for gene in rs3val["Target Gene Symbol"].unique():
    #all sgRNAs that target that gene 
    gene_sgRNAs=rs3val[rs3val["Target Gene Symbol"]==gene]["Gene and Sequence"].tolist()
    #select 5 of these sgRNAs randomly
    if len(gene_sgRNAs)>=5: #one gene only targeted by 4 sgRNAs. exclude. 
        random_5=random.sample(gene_sgRNAs,5)
    sgRNAs_selected=sgRNAs_selected+random_5
random_library=rs3val[rs3val["Gene and Sequence"].isin(sgRNAs_selected)].copy()
random_library
```

```python
random_library_noness=random_library[random_library["Category"]=="Non-essential"].reset_index(drop=True).copy()
perc_prom_random=100*len(random_library_noness[random_library_noness['z_scored_sgRNA_lfc']<-2])/len(random_library_noness)
perc_active_random=100*len(random_library[random_library["On-Target Rank"]<=5])/len(random_library)

```

```python
(fig,ax)= plt.subplots(1,2,figsize=(12,5))
fig.suptitle("Effect of off-target avoidance strategy on library composition (5 sgRNA/gene)",fontsize=16)

ax[0].plot(thresholds,aggcfd_percent_prom_list,color="royalblue",label="Agg. CFD",linewidth=4)
ax[0].axhline(y=perc_prom_random,label="Random",color="greenyellow",linewidth=4)
ax[0].axhline(y=perc_prom_rank,label="Off-Target Rank",color="skyblue",linewidth=4)
ax[0].axhline(y=perc_prom_guidescan,label="Guidescan",color="teal",linewidth=4)
#replot to show on top
ax[0].plot(thresholds,aggcfd_percent_prom_list,color="royalblue",linewidth=4)
ax[0].set_xlabel("Agg. CFD threshold",fontsize=16)
ax[0].set_ylabel("% of sgRNAs (nonessentials only)",fontsize=16)
ax[0].set_ylim(0,100)
ax[0].set_xlim(0.1,4.9)
ax[0].legend(fontsize=16,loc="upper left")
ax[0].set_title("Promiscuous in DeWeirdt 2022 screen",fontsize=16)


ax[1].plot(thresholds,aggcfd_percent_active_list,color="royalblue",label="Agg. CFD",linewidth=4)
ax[1].axhline(y=perc_active_random,label="Random",color="greenyellow",linewidth=4)
ax[1].axhline(y=perc_active_rank,label="Off-Target Rank",color="skyblue",linewidth=4)
ax[1].axhline(y=perc_active_guidescan,label="Guidescan",color="teal",linewidth=4)
#replot to show on top
ax[1].plot(thresholds,aggcfd_percent_active_list,color="royalblue",linewidth=4)
ax[1].set_xlabel("Agg. CFD threshold",fontsize=16)
ax[1].set_ylim(0,100)
ax[1].set_xlim(0.1,4.9)
ax[1].set_ylabel("% of sgRNAs in library",fontsize=16)
ax[1].set_title("In top 5 RS3 scores for gene",fontsize=16)

plt.savefig("../Figures/sliding_aggcfd_threshold_librarymetrics.png",bbox_inches="tight")

```
