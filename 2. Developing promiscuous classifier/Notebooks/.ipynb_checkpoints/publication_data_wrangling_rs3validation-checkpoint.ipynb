### Data Cleaning

```python
#importing plate 1 and 2 and merging them
reads_df = pd.read_excel("../Data/deweirdt_2022_suppdata4.xlsx",sheet_name="Chen_2013_tracr_read_counts",skiprows=[0,1])
reads_df
```

```python
def lognorm(reads):
    """
    Standardize read counts by calculating reads per million,
    adding a pseudo-count of one, and taking the log2

    :param reads: numpy or pandas array
    :returns: numpy or pandas array
    """
    reads_per_million = (reads/reads.sum())*(10**6)
    lognormed_reads = np.log2(reads_per_million + 1)
    return lognormed_reads
condition_columns = ['pDNA', 'RepA', 'RepB',
                     'RepC']
lognorm_df = reads_df.copy()
lognorm_df[condition_columns] = lognorm_df[condition_columns].apply(lognorm)
condition_columns.remove("pDNA")
lognorm_df
```

```python
#remove underrepresented sgRNAs
filtered_lognorms = lognorm_df.copy()
z_low = -3  # minimum z-score
# Z-score the pDNA column
filtered_lognorms["pDNA_zscore"] = (filtered_lognorms['pDNA'] - filtered_lognorms['pDNA'].mean())/filtered_lognorms['pDNA'].std()
# Filter by z-score
filtered_lognorms = filtered_lognorms[filtered_lognorms["pDNA_zscore"] > z_low].reset_index(drop=True)
# Drop z-scored columns
filtered_lognorms = filtered_lognorms.drop("pDNA_zscore", axis=1)
filtered_lognorms
```

```python
lfc_df = filtered_lognorms.copy()
for col in condition_columns:
    lfc_df[col] = filtered_lognorms[col] - filtered_lognorms["pDNA"]
lfc_df=lfc_df.drop("pDNA",axis=1)
lfc_df
```

```python
#check replicate correlation
lfc_correlations = lfc_df.corr()
sns.clustermap(lfc_correlations, cmap='RdBu_r',
               vmin=-1, vmax=1, annot=True)
```

```python
#adequate replicate correlation: take average
lfc_df["sgRNA_lfc"]=lfc_df.mean(axis=1)
lfc_df=lfc_df.drop(condition_columns,axis=1)
lfc_df
```

```python
#merge with annotations

annotations_df = pd.read_excel("../Data/deweirdt_2022_suppdata4.xlsx",sheet_name="Library annotation",usecols=["sgRNA Sequence","Target Gene Symbol"])
annotations_df
```

```python
annotated_sgrna_lfcs = lfc_df.merge(annotations_df,how='inner',on='sgRNA Sequence')
annotated_sgrna_lfcs
```

Standardize log fold changes to the intergenic controls, which represent viability loss from double stranded break. Doing so isolates viability effect attributed to gene function (does not exclude off-target effects, however).

This may be good to demonstrate in the the presentation but I don't think we should test effect of not doing this (moreso enhances individual log fold change interpretation).

```python

control_sgrna_lfcs = (annotated_sgrna_lfcs[(annotated_sgrna_lfcs['Target Gene Symbol']
                                            .str.contains('INTERGENIC'))]
                      .reset_index(drop=True))
control_sgrna_mean=control_sgrna_lfcs["sgRNA_lfc"].mean(axis=0)
control_sgrna_sd=control_sgrna_lfcs["sgRNA_lfc"].std(axis=0,numeric_only=True)
zscored_sgrna_lfcs=annotated_sgrna_lfcs.copy()
zscored_sgrna_lfcs['z_scored_sgRNA_lfc'] = (zscored_sgrna_lfcs['sgRNA_lfc'] - control_sgrna_mean)/control_sgrna_sd
zscored_sgrna_lfcs
```

### Standard analysis: gene-level z-scores

To combine z-scores at the gene level, we sum the sgRNA z-scores and divide by the square root of the number of sgRNAs. Doing so ensures that the individual score of each sgRNA targeting a gene contributes evenly to the gene-level score.



```python
zscored_gene_lfcs = (zscored_sgrna_lfcs.groupby('Target Gene Symbol')
                     .agg(z_scored_lfc_sum = ('z_scored_sgRNA_lfc', 'sum'),
                          n_sgrnas = ('z_scored_sgRNA_lfc', 'count'))
                     .reset_index())
zscored_gene_lfcs['z_scored_lfc'] = (zscored_gene_lfcs['z_scored_lfc_sum']/np.sqrt(zscored_gene_lfcs['n_sgrnas']))
zscored_gene_lfcs = zscored_gene_lfcs.drop('z_scored_lfc_sum', axis=1)
zscored_gene_lfcs
```

To assess how well the z-scores of each gene from the experimental data reflect the true effect of its loss on viablity, I am assessing the position of gold-standard essential genes along this list. 

```python
#rank genes by zscore: rank of 1 is lowest z score (largest viability decrease)
zscored_gene_lfcs["z_score_rank"]=zscored_gene_lfcs["z_scored_lfc"].rank()
zscored_gene_lfcs
```

```python
#layer in info re: which genes are known to be essential
essential_genes=pd.read_csv("../Data/essential_gene.csv")
essential_genes=essential_genes["Target Gene Symbol"].tolist()
zscored_gene_lfcs["known_essential_gene"]=zscored_gene_lfcs["Target Gene Symbol"].isin(essential_genes)
zscored_gene_lfcs
```

```python
groups = zscored_gene_lfcs.groupby('known_essential_gene')
for name, group in groups:
    plt.plot(group["z_score_rank"], group["z_scored_lfc"], marker='o', linestyle='', label=name)

plt.xlabel("Perturbation rank (1= greatest viability loss)")
plt.ylabel("Perturbation Z-score")
plt.legend(title="Targeting Essential Gene")
```

Most of the genes with the highest z-scores are known essentials.

Of the top 10% of sensitizing hits, what % are truly essential? (Essentials make up under 10% of genes targeted, so not expected to be 100%)

```python
top_10th_genes=zscored_sgrna_lfcs[zscored_sgrna_lfcs["z_score_rank"]<=round(len(zscored_gene_lfcs)/10)].reset_index(drop=True).copy()
len(top_10th_genes[top_10th_genes["known_essential_gene"]==True])/len(top_10th_genes)
```
