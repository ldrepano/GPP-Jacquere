```python
import pandas as pd
import numpy as np
from sklearn.metrics import f1_score
import gzip
import matplotlib.pyplot as plt
import seaborn as sns
import gpplot as gpp
gpp.set_aesthetics(context = 'paper')
```

Using RS3 validation set nonessential-targeting guides (Deweirdt et al 2022) screened with CRISRko/Cas9 + Chen tracr, evaluating the performance of various metrics to classify guides as promiscuous from predicted off-target sites. 


Classifiers to compare:
1. **Bin I Match I Count >3** 

This is the picking round criteria currently implemented in CRISPick. Guides with more than three off-target sites with a CFD score of 1 are considered promiscuous.


2. **Guidescan** 

Guidescan defines a promiscuous guide as having:

- any OTS with 0 mismatches to target site

- more than 1 OTS with 1 mismatch to target site 

- CFD sum of OTS with up to 3 mismatches of >=4 (this is equivalent to specificity score 0.2) 

and includes no OTS with an NGG PAM in this search. Mismatch # includes any mismatch in entire 20mer.

3. **Aggregate CFD**

Variation of CFD sum employed in Guidescan. The set of off-target sites included in this sum as well as the threshold to classify a guide as promiscuous are empirically determined in this notebook.

## Loading data

Load screening data from prior notebook that indicates whether or not each nonessential-targeting guide is promiscuous. Note that this data excludes nonessential-targeting guides that are between 1 and 2 standard deviations below the mean of intergenic controls (Z-score between -1 and -2) since it is unclear if these guides have substantial off-target activity.

```python
screen_data= pd.read_csv("../Data/rs3validation_nonessentials_zscored.csv",index_col=None)
screen_data
```

Acquiring the RS3 sequence scores of guides, which indicates the expected activity of the guide on the basis of its sequence features. This was acquired by running the cleaned RS3 validation set nonessentials through CRISPick with RS3Seq-Chen (no target) as the on-target scorer. 

```python
RS3seq_scored= pd.read_table("../Data/rs3val_nonessentialscleaned_scoredrs3seq_notarget-sgrna-designs.txt",usecols=["sgRNA Sequence","On-Target Ruleset","On-Target Efficacy Score"])
RS3seq_scored["high_RS3seq_score"]= RS3seq_scored["On-Target Efficacy Score"]>=0.2
RS3seq_scored
```

Below are the results of running a developmental version of CRISPick (where the output file show off-target sites (OTS) for each sgRNA on their own line, detailing mismatch number and CFD). The input of this CRISPick run is the nonessential-targeting guides for the RS3 Validation set and Avana Library, thus not all of the data loaded below is used in subsequent analysis. 


```python
with gzip.open('../Data/sgrna-designs.targetdisco.txt.gz') as f:
    crispick_verbose = pd.read_table(f,index_col=False,header=None,names=["Target","Target Sequence","Context Sequence","Match Tier", "CFD Score","PAM Sequence", "# of mismatches"])
crispick_verbose
```

Using this data to calculate both aggregate CFD and #Bin I OTS

```python
#get rid of NA pick orders (i.e. MAX OTS)
crispick_verbose=crispick_verbose[pd.to_numeric(crispick_verbose["# of mismatches"], errors='coerce').notnull()].reset_index(drop=True).copy()
crispick_verbose["# of mismatches"]=crispick_verbose["# of mismatches"].astype(np.int64)

#calculate Agg CFD on up to 0,1,2,3 MM
crispick_verbose["upto0"]= np.where(crispick_verbose["# of mismatches"]==0, crispick_verbose["CFD Score"].astype(np.float64),0)
crispick_verbose["upto1"]= np.where(crispick_verbose["# of mismatches"]<=1, crispick_verbose["CFD Score"].astype(np.float64),0)
crispick_verbose["upto2"]= np.where(crispick_verbose["# of mismatches"]<=2, crispick_verbose["CFD Score"].astype(np.float64),0)
crispick_verbose["upto3"]= np.where(crispick_verbose["# of mismatches"]<=3, crispick_verbose["CFD Score"].astype(np.float64),0)

#also get # Bin I OTS
crispick_verbose["is_BinI"]= np.where(crispick_verbose["CFD Score"]==1, 1,0)


crispick_agg = (crispick_verbose.groupby(["Target","Target Sequence"])
            .agg(upto0= ("upto0","sum"),
                 upto1= ("upto1","sum"),
                 upto2= ("upto2","sum"),
                 upto3= ("upto3","sum"),
                 binIsites= ("is_BinI","sum")
                )
            .reset_index())
crispick_agg
```



The RS3 validation set was evaluated with the [Guidescan2 gRNA design tool](https://guidescan.com/) to acquire specificity scores

```python
guidescan_batch1=pd.read_csv("../Data/rs3val_ess_noness_guidescan_batch1.csv")
guidescan_batch2=pd.read_csv("../Data/rs3val_ess_noness_guidescan_batch2.csv")
guidescan_batch3=pd.read_csv("../Data/rs3val_ess_noness_guidescan_batch3.csv")

guidescan=pd.concat([guidescan_batch1,guidescan_batch2]).reset_index(drop=True)
guidescan=pd.concat([guidescan,guidescan_batch3]).reset_index(drop=True)
```

The most recent [update](https://www.biorxiv.org/content/10.1101/2022.05.02.490368v1) from Guidescan2 developers suggests a specificity score of 0.2 as a cutoff for promiscuous guides. 

```python
#specificity=0 if it violates the 0 or 1mm rule (therefore is 'not in database')
guidescan["guidescan_specificity"]=np.where(guidescan["Chromosome"].str.contains("chr"), guidescan["Specificity"],0)

guidescan_cutoff= 0.2
guidescan["guidescan_below_0.2"]=guidescan["guidescan_specificity"].astype(np.float64)<guidescan_cutoff

guidescan["sgRNA Sequence"]=guidescan["gRNA"].str[:20]


guidescan
```

```python
guidescan["Chromosome"].value_counts()
```

Merging all dataframes to have CRISPick, Guidescan, and screening data for each guide in one dataframe.

```python
rs3val_agg=crispick_agg.merge(screen_data[["On-target Gene Symbols","sgRNA Sequence","promiscuous"]],left_on="Target Sequence",right_on="sgRNA Sequence")
rs3val_agg=rs3val_agg.merge(guidescan[["sgRNA Sequence","Chromosome","guidescan_specificity","guidescan_below_0.2"]],on="sgRNA Sequence")
rs3val_agg=rs3val_agg.merge(RS3seq_scored[["sgRNA Sequence","high_RS3seq_score"]],on="sgRNA Sequence")
rs3val_agg=rs3val_agg[["sgRNA Sequence","high_RS3seq_score","Chromosome","promiscuous","upto0","upto1","upto2","upto3","binIsites","guidescan_specificity","guidescan_below_0.2"]]
rs3val_agg
```

CRISPick and Guidescan yield duplicate guides/off-target sites due to the fact that various guides target multiple recognized genes. Since all instances of these guides in the dataframe have the same off-target profile, keeping any copy of the guide is valid for the purpose of identifying promiscuous guides

```python
rs3val_agg=rs3val_agg.drop_duplicates(subset="sgRNA Sequence",keep="first").reset_index(drop=True)
```

```python
#Counts of guides for which data is available in CRISPick, Guidescan (decrease likely due to annotation drift)
print("total # guides for training",len(rs3val_agg))
print("total # promiscuous guides",len(rs3val_agg[rs3val_agg["promiscuous"]==True]))
print("total # nonpromiscuous guides",len(rs3val_agg[rs3val_agg["promiscuous"]==False]))
```

```python
rs3val_agg["Chromosome"].value_counts()
```

# Section 1: Identifying # of mismatches to include in Aggregate CFD

Aggregate CFD refers to profiling each guide by the sum of CFD scores (predicted probability of off-target activity) of putative off-target sites. Building a classifier on Aggregate CFD requires (1) determining which putative off-target sites should be included in the CFD sum, then (2) identify an aggregate CFD threshold that most effectively separates promiscuous from nonpromiscuous guides.


*Identifying the ideal breadth of off-target sites to include in the calculation of Aggregate CFD*. 

Evaluating Agg CFD on off-target sites with up to 1 vs 2 vs 3 mismatches in the SDR to the guide using Macro F1 Score. The SDR, or Specificity-Defining Region, for application with Cas9 counts mismatches in positions 3-20 or the latter 2 positions in the PAM. 

### Calculate F1 to compare Agg CFD with 0-3 MM OTS

The reported F1 is the best possible for that given number, iteratively testing all possible aggregate CFD cutoffs.

```python
metric=rs3val_agg["upto0"].astype(np.float64)
f1=[]
cutoffs=[i*0.1 for i in range(300)]
for cutoff in cutoffs:
    metric_binary= metric > cutoff
    performance= f1_score(rs3val_agg["promiscuous"], metric_binary ,average="macro")
    f1.append(performance)
f1_0mm_rs3val=max(f1) 
bestcutoff_0mm=cutoffs[f1.index(f1_0mm_rs3val)]
rs3val_agg["Predicted (Agg CFD up to 0MM)"]=rs3val_agg["upto0"]>bestcutoff_0mm
print("up to 0mm has f1 of ", f1_0mm_rs3val, " at Agg CFD cutoff ",bestcutoff_0mm)
```

```python
pd.DataFrame(rs3val_agg[["Predicted (Agg CFD up to 0MM)","promiscuous"]].value_counts()).sort_values(by="Predicted (Agg CFD up to 0MM)")
```

```python
metric=rs3val_agg["upto1"].astype(np.float64)
f1=[]
cutoffs=[i*0.1 for i in range(300)]
for cutoff in cutoffs:
    metric_binary= metric > cutoff
    performance= f1_score(rs3val_agg["promiscuous"], metric_binary ,average="macro")
    f1.append(performance)
f1_1mm_rs3val=max(f1) 
bestcutoff_1mm=cutoffs[f1.index(f1_1mm_rs3val)]
rs3val_agg["Predicted (Agg CFD up to 1MM)"]=rs3val_agg["upto1"]>bestcutoff_1mm
print("up to 1mm has f1 of ", f1_1mm_rs3val, " at Agg CFD cutoff ",bestcutoff_1mm)
```

```python
pd.DataFrame(rs3val_agg[["Predicted (Agg CFD up to 1MM)","promiscuous"]].value_counts()).sort_values(by="Predicted (Agg CFD up to 1MM)")
```

```python
metric=rs3val_agg["upto2"].astype(np.float64)
f1=[]
cutoffs=[i*0.1 for i in range(500)]
for cutoff in cutoffs:
    metric_binary= metric > cutoff
    performance= f1_score(rs3val_agg["promiscuous"], metric_binary ,average="macro")
    f1.append(performance)
f1_2mm_rs3val=max(f1) 
bestcutoff_2mm=cutoffs[f1.index(f1_2mm_rs3val)]
rs3val_agg["Predicted (Agg CFD up to 2MM)"]=rs3val_agg["upto2"]>bestcutoff_2mm
print("up to 2mm has f1 of ", f1_2mm_rs3val, " at Agg CFD cutoff ",bestcutoff_2mm)
```

```python
plt.title("Predicting promiscuity in RS3 validation set",size=16)

f1scores_aggcfd_mm_rs3= pd.DataFrame(data={"OTS included (up to _ MM)":["0MM","1MM","2MM","3MM"],"F1 score":[f1_0mm_rs3val,f1_1mm_rs3val,f1_2mm_rs3val,f1_3mm_rs3val]})
ax=sns.barplot(data=f1scores_aggcfd_mm_rs3,x="OTS included (up to _ MM)",y="F1 score",color="cornflowerblue")
plt.bar_label(ax.containers[0], fmt = '%0.2f',size=14)
plt.yticks([])
plt.xlabel("Off-Target Sites included in Aggregate CFD",size=16)
plt.ylabel("F1 Score",size=16)
plt.ylim(0,f1scores_aggcfd_mm_rs3["F1 score"].max()*1.1)
plt.xticks(labels=["0MM","≤1MM","≤2MM","≤3MM"],ticks=[0,1,2,3],size=16)
gpp.savefig("../Figures/agg_cfd_0_to_3_mm_f1_rs3val.pdf", dpi=300, bbox_inches = 'tight')
```

# Section 2: Compare Aggregate CFD to other potential classifiers



### Bin I Match Count >3

```python
rs3val_agg["over_3_binI_sites"]= rs3val_agg["binIsites"]>3 
f1_crispick=f1_score(rs3val_agg["promiscuous"],rs3val_agg["over_3_binI_sites"] ,average="macro")
f1_crispick
```

```python
pd.DataFrame(rs3val_agg[["over_3_binI_sites","promiscuous"]].value_counts()).sort_values(by="over_3_binI_sites")
```

### Guidescan

Performance of threshold recommended for library design, [source](https://www.biorxiv.org/content/10.1101/2022.05.02.490368v1.full.pdf)

```python
f1_guidescan=f1_score(rs3val_agg["promiscuous"],rs3val_agg["guidescan_below_0.2"] ,average="macro")
f1_guidescan
```

```python
pd.DataFrame(rs3val_agg[["guidescan_below_0.2","promiscuous"]].value_counts()).sort_values(by="guidescan_below_0.2")
```

Performance that optimizes specificity score as a promiscuous classifier

```python
metric=rs3val_agg["guidescan_specificity"].astype(np.float64)
f1=[]
cutoffs=[i*0.01 for i in range(100)]
for cutoff in cutoffs:
    metric_binary= metric < cutoff
    performance= f1_score(rs3val_agg["promiscuous"], metric_binary ,average="macro")
    f1.append(performance)
guidescan_opt_f1=max(f1) 
bestcutoff_guidescan=cutoffs[f1.index(guidescan_opt_f1)]
rs3val_agg["optimized guidescan"]=rs3val_agg["guidescan_specificity"]>bestcutoff_guidescan
print("guidescan has f1 of ", guidescan_opt_f1, " at specificity score cutoff ",bestcutoff_guidescan)
```

note this threshold is close to the 0.05 employed in visualizations in the guidescan2 paper. note below that this contains many false positives

```python
pd.DataFrame(rs3val_agg[["optimized guidescan","promiscuous"]].value_counts()).sort_values(by="optimized guidescan")
```

```python
plt.title("Predicting promiscuity in RS3 validation set",size=16)

f1scores= pd.DataFrame(data={"Classifier":["Agg CFD (<=1 MM)","CRISPick","<0.04 Specificity Guidescan ","<0.2 Specificity Guidescan"],"F1 score":[f1_1mm_rs3val,f1_crispick, guidescan_opt_f1,f1_guidescan]})
ax=sns.barplot(data=f1scores,x="Classifier",y="F1 score",color="cornflowerblue")
plt.bar_label(ax.containers[0], fmt = '%0.2f',size=14)
plt.yticks([])
plt.xlabel("Promiscuous Classifier",size=16)
plt.ylabel("F1 Score",size=16)
plt.ylim(0,f1scores["F1 score"].max()*1.1)
plt.xticks(labels=["Agg CFD \n>4.6","CRISPick","Guidescan \n<0.04","Guidescan \n<0.2"],ticks=[0,1,2,3],size=14,rotation=10)

gpp.savefig("../Figures/agg_cfd_crispick_guidescan_f1.pdf", dpi=300, bbox_inches = 'tight')
```

## Exploration: Effect of on-target activity on promiscuous predictions

Each of these classifiers include a set of false positives; guides predicted to be promiscuous based on putative off-target sites, but are not promiscuous. One possible explanation for their lack off (observed) off-target activity is that these guides are in general not prone to activity. To examine if this is the case, I am repeating the analysis with only guides expected to have strong activity on the basis of their sequence features (RS3 Seq >0.2) and examining if this improves the performance of the classifiers. This performance may be more relevant since CRISPick preferentially selects guides with high predicted activity to protect the on-target activity of the library. 


```python
rs3val_highactivity=rs3val_agg[rs3val_agg["high_RS3seq_score"]].reset_index(drop=True).copy()
print("filtering for high on target activity reduced library size from", len(rs3val_agg), "to", len(rs3val_highactivity))
```
