```python
import pandas as pd
import numpy as np
from sklearn.metrics import f1_score
import gzip
import matplotlib.pyplot as plt
import seaborn as sns
import gpplot as gpp
gpp.set_aesthetics(context = 'paper')
```

Testing new off-target metrics on example genes

## Loading data

Load screening data from prior notebook that indicates whether or not each nonessential-targeting guide is promiscuous. Note that this data excludes nonessential-targeting guides that are between 1 and 2 standard deviations below the mean of intergenic controls (Z-score between -1 and -2) since it is unclear if these guides have substantial off-target activity. This data also excludes guides with RS3 Sequence score below 0.2. 

```python
screen_data= pd.read_csv("../Data/rs3valdata_updatedmappings_zscore.csv")
screen_data
```

```python
#crispick run to get pick orders (note: RS3 seq, does not factor in target)
crispick= pd.read_table("../Data/rs3val_nonessentialscleaned_scoredrs3seq_notarget-sgrna-designs.txt",usecols=['sgRNA Sequence','Target Gene Symbol','Target Cut %',"On-Target Rank","Off-Target Rank", "Combined Rank"])
crispick
```

Below are the results of running a developmental version of CRISPick (where the output file show off-target sites (OTS) for each sgRNA on their own line, detailing mismatch number and CFD). The input of this CRISPick run is the nonessential-targeting guides for the RS3 Validation set and Avana Library, thus not all of the data loaded below is used in subsequent analysis. 


```python
with gzip.open('../Data/sgrna-designs.targetdisco.txt.gz') as f:
    crispick_verbose = pd.read_table(f,index_col=False,header=None,names=["Target","Target Sequence","Context Sequence","Match Tier", "CFD Score","PAM Sequence", "# of mismatches"])
crispick_verbose
```

Using this data to calculate both aggregate CFD and #Bin I OTS

```python
#get rid of NA pick orders (i.e. MAX OTS)
crispick_verbose=crispick_verbose[pd.to_numeric(crispick_verbose["# of mismatches"], errors='coerce').notnull()].reset_index(drop=True).copy()
crispick_verbose["# of mismatches"]=crispick_verbose["# of mismatches"].astype(np.int64)

#Agg CFD
crispick_verbose["cfd_under2mm"]= np.where(crispick_verbose["# of mismatches"]<=1, crispick_verbose["CFD Score"].astype(np.float64),0)
#influential OTS filters
crispick_verbose["perfectmatch"]= np.where((crispick_verbose["# of mismatches"]<=2)&(crispick_verbose["CFD Score"]==1), 1,0)
crispick_verbose["perfectmatch_proteincoding"]= np.where((crispick_verbose["# of mismatches"]<=2)&(crispick_verbose["CFD Score"]==1)&(crispick_verbose["Match Tier"]=="Tier I"), 1,0)

crispick_agg = (crispick_verbose.groupby(["Target","Target Sequence"])
            .agg(agg_cfd= ("cfd_under2mm","sum"),
                 perfectmatch=("perfectmatch","max"),
                 perfectmatch_proteincoding=("perfectmatch_proteincoding","max"),
                )
            .reset_index())
crispick_agg
```

Merging all dataframes to have CRISPick and screening data for each guide in one dataframe.

```python
rs3val_agg=crispick_agg.merge(screen_data[["On-target Gene Symbols","sgRNA Sequence","z_scored_sgRNA_lfc"]],left_on="Target Sequence",right_on="sgRNA Sequence")
rs3val_agg=rs3val_agg[["sgRNA Sequence","On-target Gene Symbols","agg_cfd","perfectmatch","perfectmatch_proteincoding","z_scored_sgRNA_lfc"]].merge(crispick,on="sgRNA Sequence")
rs3val_agg["predicted_promiscuous_strict"]=(rs3val_agg["agg_cfd"]>4.8)|(rs3val_agg["perfectmatch"]==1)
rs3val_agg["predicted_promiscuous_relaxed"]=(rs3val_agg["agg_cfd"]>4.8)|(rs3val_agg["perfectmatch_proteincoding"]==1)

rs3val_agg
```

Assesing rank-based selection

```python
OR9Q2_data=rs3val_agg[rs3val_agg["On-target Gene Symbols"]=="OR9Q2"].reset_index(drop=True)
plt.axhline(y=-2,label="2 S.D. below intergenic mean",c="grey",linestyle="--",linewidth=4)
plt.scatter(data=OR9Q2_data[OR9Q2_data["Combined Rank"]>OR9Q2_data["On-Target Rank"]],x="Target Cut %",y="z_scored_sgRNA_lfc",c="red",label="Penalized",linewidths=2)
plt.scatter(data=OR9Q2_data[OR9Q2_data["Combined Rank"]<=OR9Q2_data["On-Target Rank"]],x="Target Cut %",y="z_scored_sgRNA_lfc",c="cornflowerblue",label="Not penalized",linewidths=2)

plt.legend(fontsize=16,bbox_to_anchor=(1,1))
plt.xlabel("Position along protein-coding region",fontsize=16)
plt.ylabel("Z-score",fontsize=16)
plt.title("Guides penalized by off-target rank (OR9Q2)",fontsize=16)
plt.savefig("../Figures/OR9Q2_rankbased.png", dpi=600, bbox_inches = 'tight')
```

Assessing new method of guide selection

```python
OR9Q2_data=rs3val_agg[rs3val_agg["On-target Gene Symbols"]=="OR9Q2"].reset_index(drop=True)
plt.axhline(y=-2,label="2 S.D. below intergenic mean",c="grey",linestyle="--",linewidth=4)
plt.scatter(data=OR9Q2_data[OR9Q2_data["predicted_promiscuous_relaxed"]],x="Target Cut %",y="z_scored_sgRNA_lfc",c="red",label="predicted promiscuous",linewidths=2)
plt.scatter(data=OR9Q2_data[OR9Q2_data["predicted_promiscuous_relaxed"]==False],x="Target Cut %",y="z_scored_sgRNA_lfc",c="cornflowerblue",label="predicted nonpromiscuous",linewidths=2)
plt.legend(fontsize=16,bbox_to_anchor=(1,1))
plt.xlabel("Position along protein-coding region",fontsize=16)
plt.ylabel("Z-score",fontsize=16)
plt.title("New approach more precisely detects promiscuous guides (OR9Q2)",fontsize=16)
plt.savefig("../Figures/OR9Q2_new.png", dpi=600, bbox_inches = 'tight')
```
