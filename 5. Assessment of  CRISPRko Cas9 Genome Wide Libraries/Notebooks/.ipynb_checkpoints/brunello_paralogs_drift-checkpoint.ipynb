# Part 2: Behavior of "drifted" guides in Brunello screen data

```python
brunello_updated_mappings["Notes"]=brunello_updated_mappings["Notes"].fillna("Protein Coding")
brunello_updated_mappings["Notes"].value_counts()
```

Brunello Screening data [source](https://www.nature.com/articles/s41467-020-14620-6#Sec28) Supplementary Dataset 1

```python
brunello_screening_data=brunello_screening_data=pd.read_excel("../Data/41467_2020_14620_MOESM4_ESM.xlsx",sheet_name="Brunello",header=[0,1,2,3],index_col=[0,1])
brunello_A375_dropout=brunello_screening_data["A375"]["Control guide"]["Dropout"].reset_index()
brunello_A375_dropout=brunello_A375_dropout.rename(columns={"level_0": "sgRNA Sequence", "level_1": "pDNA"})
brunello_A375_dropout
```

Brunello has 1000 nontargeting control sgRNAs. These can be recognized as the guides in the screening data but not in the LIMS mapping results (since I did not include NTs in that run)

```python
nt_controls=brunello_A375_dropout[-brunello_A375_dropout["sgRNA Sequence"].isin(brunello_updated_mappings["Target Sequence"].tolist())]["sgRNA Sequence"].tolist()
```

```python
def lognorm(reads):
    """
    Standardize read counts by calculating reads per million,
    adding a pseudo-count of one, and taking the log2

    :param reads: numpy or pandas array
    :returns: numpy or pandas array
    """
    reads_per_million = (reads/reads.sum())*(10**6)
    lognormed_reads = np.log2(reads_per_million + 1)
    return lognormed_reads
condition_columns = ['pDNA', 'RepA', 'RepB']
lognorm_df = brunello_A375_dropout.copy()
lognorm_df[condition_columns] = lognorm_df[condition_columns].apply(lognorm)
condition_columns.remove("pDNA")

#remove underrepresented sgRNAs
filtered_lognorms = lognorm_df.copy()
z_low = -3  # minimum z-score
# Z-score the pDNA column
filtered_lognorms["pDNA_zscore"] = (filtered_lognorms['pDNA'] - filtered_lognorms['pDNA'].mean())/filtered_lognorms['pDNA'].std()
# Filter by z-score
filtered_lognorms = filtered_lognorms[filtered_lognorms["pDNA_zscore"] > z_low].reset_index(drop=True)
# Drop z-scored columns
filtered_lognorms = filtered_lognorms.drop("pDNA_zscore", axis=1)
#calculate log fold change wrt pdna
lfc_df = filtered_lognorms.copy()
for col in condition_columns:
    lfc_df[col] = filtered_lognorms[col] - filtered_lognorms["pDNA"]
lfc_df=lfc_df.drop("pDNA",axis=1)
#average LFC across replicates
lfc_df["sgRNA_lfc"]=lfc_df.mean(axis=1,numeric_only=True)
lfc_df=lfc_df.drop(condition_columns,axis=1)
lfc_df
```

```python
control_sgrna_lfcs = lfc_df[(lfc_df['sgRNA Sequence'].isin(nt_controls))].reset_index(drop=True)

#verifying that controls are normally distributed with D'Agostino-Pearson K2 test
stat, p = normaltest(control_sgrna_lfcs["sgRNA_lfc"])
print('Statistics=%.3f, p=%.3f' % (stat, p),"length=",len(control_sgrna_lfcs))

#z-score guides relative to NT controls
control_sgrna_mean=control_sgrna_lfcs["sgRNA_lfc"].mean(axis=0)
control_sgrna_sd=control_sgrna_lfcs["sgRNA_lfc"].std(axis=0,numeric_only=True)
zscored_sgrna_lfcs=lfc_df.copy()
zscored_sgrna_lfcs['z_scored_sgRNA_lfc'] = (zscored_sgrna_lfcs['sgRNA_lfc'] - control_sgrna_mean)/control_sgrna_sd
zscored_sgrna_lfcs
```

Associate guides with original mappings

```python
brunello_og_mapping=pd.read_excel("../Data/Original Annotations/STable 21 Brunello.xlsx",sheet_name="Sheet1")
zscored_sgrna_lfcs_annotated=zscored_sgrna_lfcs.merge(brunello_og_mapping[["sgRNA Target Sequence", "Target Gene Symbol"]],left_on="sgRNA Sequence",right_on="sgRNA Target Sequence")
zscored_sgrna_lfcs_annotated                         
                         
```

```python
brunello_data_drift=zscored_sgrna_lfcs_annotated.merge(brunello_updated_mappings[["Target Sequence","Notes"]],left_on="sgRNA Sequence",right_on="Target Sequence",how="left")
brunello_data_drift=brunello_data_drift[-brunello_data_drift["Notes"].isna()].reset_index(drop=True)
brunello_data_drift
```

```python
#identify which are essential
essential_genes=pd.read_csv("/Users/ldrepano/Library/CloudStorage/GoogleDrive-ldrepano@broadinstitute.org/Shared drives/GPP Cloud /R&D/People/Laura/CRISPick Parameter Tuning/Manuscript/Part 2: Developing promiscuous classifier/Data/AchillesCommonEssentialControls.csv")
essential_genes["Gene"]=essential_genes["Gene"].apply(lambda x: x.split("(")[0])
#remove space from gene name 
essential_genes["Gene"]=essential_genes["Gene"].str.replace(" ","")
essential_gene_list=essential_genes["Gene"].tolist()

essential_targeting_brunello=brunello_data_drift[brunello_data_drift["Target Gene Symbol"].isin(essential_gene_list)].reset_index(drop=True)
essential_targeting_brunello
```
