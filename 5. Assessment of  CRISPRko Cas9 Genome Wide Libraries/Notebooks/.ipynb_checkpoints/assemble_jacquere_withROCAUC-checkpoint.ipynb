```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import roc_curve, auc
import seaborn as sns
import random
```

Combining first set of CRISPick runs to assemble Jacquere in order to ...
- Assess the guides selected from the picking scheme, troubleshooting any odd patterns observed
- Assess if 4 guides per gene is necessary, or if this improved picking scheme enables a more compact library

```python
#CRISPick output of three design runs: 
## 1. all protein coding genes in ensembl
## 2. those in NCBI but not ensembl
## 3. those unique to CHESS
ensembl_designs=pd.read_table("../Jacquere_library/ensembl_113_protein_coding_aggcfd_targetlocal-sgrna-designs.txt")
NCBI_designs=pd.read_table("../Jacquere_library/NCBI_not_ensembl_protein_coding_genes_aggcfd_targetlocal-sgrna-designs.txt")
CHESS_designs= pd.read_table("../Jacquere_library/chess_targetlocal_aggcfd-sgrna-designs.txt")

#replace exon ranges with identifiers for CHESS genes 
CHESS_identifier= pd.read_csv("../Jacquere_library/design_annotations/CHESS3.1.3_unique_protein_coding_genes_coordinates_with_identifier.csv",header=None)
CHESS_designs=CHESS_designs.merge(CHESS_identifier,left_on="Input",right_on=1)
CHESS_designs["Input"]=CHESS_designs[0]
CHESS_designs=CHESS_designs.drop([0,1],axis=1)

jacquere=pd.concat([ensembl_designs,NCBI_designs,CHESS_designs],keys=["Ensembl","NCBI","CHESS"]).reset_index()
jacquere["Source"]=jacquere["level_0"]
jacquere=jacquere.drop(["level_0","level_1"],axis=1)

#remove any guides with Aggregate CFD > 4.8
jacquere=jacquere[jacquere["Aggregate CFD Score"]!="MAX"].reset_index(drop=True)

jacquere
```

Identifying genes that have distinct Ensembl Gene IDs, but the same Gene Symbol and are targeted by the same set of guides. These duplicates can be removed 

```python
#get all sgRNAs (up to 4) associated with a given target
guides_per_input=pd.DataFrame(jacquere.groupby(['Input','Target Gene Symbol','Source','Chromosome'])['sgRNA Sequence'].apply(list)).reset_index()
#get sets of guides that are identical across different targets
guideset_counts= pd.DataFrame(guides_per_input["sgRNA Sequence"].value_counts()).reset_index()
repeated_guide_sets= guideset_counts[guideset_counts["count"]>1]["sgRNA Sequence"].tolist()
#all targets that share a sgRNA set with another input
targets_with_repeated_guidesets=guides_per_input[guides_per_input["sgRNA Sequence"].isin(repeated_guide_sets)].sort_values(by="sgRNA Sequence").reset_index(drop=True)

#of targets that share the same set of sgRNAs, identify those that also have the same gene symbol 
counts_per_genename_repeated_guidesets=pd.DataFrame(targets_with_repeated_guidesets["Target Gene Symbol"].value_counts()).reset_index()
repeated_genename_repeated_guidesets=counts_per_genename_repeated_guidesets[counts_per_genename_repeated_guidesets["count"]>1]["Target Gene Symbol"].tolist()
same_genename_same_guideset=guides_per_input[guides_per_input["Target Gene Symbol"].isin(repeated_genename_repeated_guidesets)].sort_values(by="Input")
#list targets that are repeated from prior targets for the same gene 
repeated_target_inputs=same_genename_same_guideset.drop_duplicates(subset="Target Gene Symbol",keep="last")["Input"].tolist()

#remove these repeats from jacquere
jacquere=jacquere[jacquere["Input"].isin(repeated_target_inputs)==False].reset_index(drop=True)
print("# repeated inputs deleted from jacquere:", len(repeated_target_inputs))

```

## Determining Ideal Quota for Jacquere

Subsampling analysis, identifying how top 1 vs 2 vs 3 vs 4 guides in Jacquere are able to yield gene z-scores appropriate for their predicted viability effect (essential vs nonessential) in RS3 validation set 

```python
rs3val_zscores= pd.read_csv("../../Part 2: Developing promiscuous classifier/Data/rs3valdata_updatedmappings_zscore.csv")
rs3val_zscores
```

```python
#get essential vs nonessential classifications 

noness_genes = pd.read_csv("../../Part 2: Developing promiscuous classifier/Data/AchillesNonessentialControls.csv")
noness_genes["Gene"]=noness_genes["Gene"].apply(lambda x: x.split("(")[0])
#remove space from gene name 
noness_genes["Gene"]=noness_genes["Gene"].str.replace(" ","")
noness_gene_list=noness_genes["Gene"].tolist()

essential_genes = pd.read_csv("../../Part 2: Developing promiscuous classifier/Data/AchillesCommonEssentialControls.csv")
essential_genes["Gene"]=essential_genes["Gene"].apply(lambda x: x.split("(")[0])
#remove space from gene name 
essential_genes["Gene"]=essential_genes["Gene"].str.replace(" ","")
essential_gene_list=essential_genes["Gene"].tolist()

rs3val_zscores["is_essential"]=rs3val_zscores["On-target Gene Symbols"].isin(essential_gene_list)
rs3val_zscores["is_nonessential"]=rs3val_zscores["On-target Gene Symbols"].isin(noness_gene_list)

print("# essential:",len(rs3val_zscores[rs3val_zscores["is_essential"]]))
print("# nonessential:",len(rs3val_zscores[rs3val_zscores["is_nonessential"]]))
```

```python
jacquere_with_z_scores=pd.merge(left=jacquere,right=rs3val_zscores,on="sgRNA Sequence")

jacquere_top1=jacquere_with_z_scores[jacquere_with_z_scores["Pick Order"]==1].reset_index(drop=True)
jacquere_top2=jacquere_with_z_scores[jacquere_with_z_scores["Pick Order"].isin([1,2])].reset_index(drop=True)
jacquere_top3=jacquere_with_z_scores[jacquere_with_z_scores["Pick Order"].isin([1,2,3])].reset_index(drop=True)
jacquere_top4=jacquere_with_z_scores
```

```python
jacquere_top4
```

Do ROC-AUC analysis on gene-level z-scores to assess optimal quota

```python
#Using stouffer's method to get gene-level z-score, separately for 1 vs 2 vs 3 vs 4 guides per gene in Jacquere

zscored_gene_lfcs_top1 = (jacquere_top1.groupby(['Target Gene Symbol',"is_essential","is_nonessential"])
                     .agg(z_score_sum = ('z_scored_sgRNA_lfc', 'sum'),
                          n_sgrnas = ('z_scored_sgRNA_lfc', 'count'))
                     .reset_index())
zscored_gene_lfcs_top1['gene_level_z_score'] = (zscored_gene_lfcs_top1['z_score_sum']/np.sqrt(zscored_gene_lfcs_top1['n_sgrnas']))
zscored_gene_lfcs_top1 = zscored_gene_lfcs_top1.drop('z_score_sum', axis=1)

zscored_gene_lfcs_top2 = (jacquere_top2.groupby(['Target Gene Symbol',"is_essential","is_nonessential"])
                     .agg(z_score_sum = ('z_scored_sgRNA_lfc', 'sum'),
                          n_sgrnas = ('z_scored_sgRNA_lfc', 'count'))
                     .reset_index())
zscored_gene_lfcs_top2['gene_level_z_score'] = (zscored_gene_lfcs_top2['z_score_sum']/np.sqrt(zscored_gene_lfcs_top2['n_sgrnas']))
zscored_gene_lfcs_top2 = zscored_gene_lfcs_top2.drop('z_score_sum', axis=1)
#exclude any genes that end up with fewer than 2 guides per gene
zscored_gene_lfcs_top2=zscored_gene_lfcs_top2[zscored_gene_lfcs_top2["n_sgrnas"]==2].reset_index()

zscored_gene_lfcs_top3 = (jacquere_top3.groupby(['Target Gene Symbol',"is_essential","is_nonessential"])
                     .agg(z_score_sum = ('z_scored_sgRNA_lfc', 'sum'),
                          n_sgrnas = ('z_scored_sgRNA_lfc', 'count'))
                     .reset_index())
zscored_gene_lfcs_top3['gene_level_z_score'] = (zscored_gene_lfcs_top3['z_score_sum']/np.sqrt(zscored_gene_lfcs_top3['n_sgrnas']))
zscored_gene_lfcs_top3 = zscored_gene_lfcs_top3.drop('z_score_sum', axis=1)
zscored_gene_lfcs_top3=zscored_gene_lfcs_top3[zscored_gene_lfcs_top3["n_sgrnas"]==3].reset_index()

zscored_gene_lfcs_top4 = (jacquere_top4.groupby(['Target Gene Symbol',"is_essential","is_nonessential"])
                     .agg(z_score_sum = ('z_scored_sgRNA_lfc', 'sum'),
                          n_sgrnas = ('z_scored_sgRNA_lfc', 'count'))
                     .reset_index())
zscored_gene_lfcs_top4['gene_level_z_score'] = (zscored_gene_lfcs_top4['z_score_sum']/np.sqrt(zscored_gene_lfcs_top4['n_sgrnas']))
zscored_gene_lfcs_top4 = zscored_gene_lfcs_top4.drop('z_score_sum', axis=1)
zscored_gene_lfcs_top4=zscored_gene_lfcs_top4[zscored_gene_lfcs_top4["n_sgrnas"]==4].reset_index()
```

```python
#contrast to random
random.seed(1)
sgRNAs_selected=[]
for gene in rs3val_zscores["On-target Gene Symbols"].unique():
    #all sgRNAs that target that gene 
    gene_sgRNAs=rs3val_zscores[rs3val_zscores["On-target Gene Symbols"]==gene]["sgRNA Sequence"].tolist()
    #select 4 of these sgRNAs randomly
    if len(gene_sgRNAs)>=4:  
        random_4=random.sample(gene_sgRNAs,4)
    sgRNAs_selected=sgRNAs_selected+random_4
random_library=rs3val_zscores[rs3val_zscores["sgRNA Sequence"].isin(sgRNAs_selected)].reset_index(drop=True).copy()

zscored_gene_lfcs_random = (random_library.groupby(['On-target Gene Symbols',"is_essential","is_nonessential"])
                     .agg(z_score_sum = ('z_scored_sgRNA_lfc', 'sum'),
                          n_sgrnas = ('z_scored_sgRNA_lfc', 'count'))
                     .reset_index())
zscored_gene_lfcs_random['gene_level_z_score'] = (zscored_gene_lfcs_random['z_score_sum']/np.sqrt(zscored_gene_lfcs_random['n_sgrnas']))
zscored_gene_lfcs_random = zscored_gene_lfcs_random.drop('z_score_sum', axis=1)
zscored_gene_lfcs_random=zscored_gene_lfcs_random[zscored_gene_lfcs_random["n_sgrnas"]==4].reset_index()
```

```python
roc_df =  pd.concat([zscored_gene_lfcs_top1,zscored_gene_lfcs_top2,zscored_gene_lfcs_top3,zscored_gene_lfcs_top4,zscored_gene_lfcs_random], keys=["Best 1 Jacquere","Best 2 Jacquere","Best 3 Jacquere","Best 4 Jacquere","Random 4"]).reset_index()
roc_df = (roc_df[roc_df['is_essential'] | roc_df['is_nonessential']]
          .reset_index(drop=True))
tpr_fpr_df_list = []
roc_auc_list = []
for group, df in roc_df.groupby('level_0'):
    fpr, tpr, treshold = roc_curve(df['is_essential'], -df['gene_level_z_score'])
    group_tpr_fpr_df = pd.DataFrame({'Frac. Essentials with \ngene-level Z below threshold': tpr, 'Frac. Nonessentials with \ngene-level Z below threshold': fpr, 'threshold': treshold})
    group_tpr_fpr_df['Guides picked per gene'] = group
    tpr_fpr_df_list.append(group_tpr_fpr_df)
    roc_auc = round(auc(fpr, tpr),3)
    roc_auc_list.append({'Guides picked per gene': group, 'ROC-AUC': roc_auc})
sgrna_roc_aucs = pd.DataFrame(roc_auc_list)
sgrna_tpr_fpr_df = (pd.concat(tpr_fpr_df_list).reset_index(drop=True))

sgrna_tpr_fpr_df_withAUC=sgrna_tpr_fpr_df.merge(sgrna_roc_aucs,on="Guides picked per gene")
sgrna_tpr_fpr_df_withAUC["Guides picked per gene"]=sgrna_tpr_fpr_df_withAUC.apply(lambda x: x["Guides picked per gene"]+" (AUC="+str(x["ROC-AUC"])+")",axis=1)
sgrna_tpr_fpr_df_withAUC
```

```python
plt.subplots(figsize=(4, 4))
sns.lineplot(data=sgrna_tpr_fpr_df_withAUC, x='Frac. Nonessentials with \ngene-level Z below threshold',
             y='Frac. Essentials with \ngene-level Z below threshold', hue='Guides picked per gene', errorbar=None)
plt.title("ROC in RS3 Validation Tiling (A375, DeWeirdt 2022)")
plt.legend(bbox_to_anchor=(1,1),title="Guides picked per gene")
sns.despine()
plt.savefig("../Figures/Jacquere_upto4guides_ROC.pdf",bbox_inches="tight")
```
