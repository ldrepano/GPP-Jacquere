Identify which guides are nonessential-targeting. This is complicated by the fact that, with updated mappings, some guides are reported to target multiple genes. Nonessential guides are defined as those whose list of targeting genes includes at least one nonessential gene and does not include any essential genes. 

To identify nonessential genes in the data, the nonessential list is dowloaded from [depmap](https://depmap.org/portal/data_page/?tab=allData&releasename=DepMap+Public+24Q2&filename=AchillesNonessentialControls.csv) 24Q2. This is the Hart 2014 list used to define nonessentials in DeWeirdt 2022.

[Source](https://depmap.org/portal/data_page/?tab=allData&releasename=DepMap+Public+24Q2&filename=AchillesCommonEssentialControls.csv) for essential-targeting genes: overlap between Biomen 2014 and Hart 2015

```python
noness_genes = pd.read_csv("../Data/AchillesNonessentialControls.csv")
noness_genes["Gene"]=noness_genes["Gene"].apply(lambda x: x.split("(")[0])
#remove space from gene name 
noness_genes["Gene"]=noness_genes["Gene"].str.replace(" ","")
noness_gene_list=noness_genes["Gene"].tolist()

essential_genes = pd.read_csv("../Data/AchillesCommonEssentialControls.csv")
essential_genes["Gene"]=essential_genes["Gene"].apply(lambda x: x.split("(")[0])
#remove space from gene name 
essential_genes["Gene"]=essential_genes["Gene"].str.replace(" ","")
essential_gene_list=essential_genes["Gene"].tolist()
```

```python
'''
Function that identifies if each guide targets a nonessential gene and,
if multiple genes are listed, that none of them are essential 
'''
def check_noness(row):
    #separate the list of genes targeted by the guide (although most guides only target 1)
    gene_targeting_list= row["On-target Gene Symbols"].split(",")
    #variable to indicate whether or not we call this guide nonessential-targeting
    row.noness_targeting=False
    #track all genes targeted by guide that are noness
    noness_genes_targeted=[]
    #check each gene that the guide targets 
    for gene in gene_targeting_list:
        #guide must target a recognized nonessential-targeting gene to be deemed nonessential-targeting
        if noness_gene_list.count(gene)>0:
            row.noness_targeting=True
            noness_genes_targeted.append(gene)
        #guide is not nonessential-targeting if any of the genes it targets is essential 
        elif essential_gene_list.count(gene)>0:
            row.noness_targeting=False
    #in the case that this gene lists multiple targeting genes, replace with list of noness-targeting
    if row.noness_targeting:
        row["On-target Gene Symbols"]=",".join(noness_genes_targeted)
    return row       
updated_mappings["On-target Gene Symbols"]=updated_mappings["On-target Gene Symbols"].astype(str)   
updated_mappings["noness_targeting"]=updated_mappings["Notes"] #this just instantiates the column 
updated_mappings=updated_mappings.apply(check_noness,axis=1)
```

```python
tkov3_reads=tkov3_reads.merge(updated_mappings[["Target Sequence","On-target Gene Symbols","Notes","noness_targeting"]],left_on="SEQUENCE",right_on="Target Sequence")
tkov3_reads
```

```python
#log normalize read counts
def lognorm(reads):
    """
    Standardize read counts by calculating reads per million,
    adding a pseudo-count of one, and taking the log2

    :param reads: numpy or pandas array
    :returns: numpy or pandas array
    """
    reads_per_million = (reads/reads.sum())*(10**6)
    lognormed_reads = np.log2(reads_per_million + 1)
    return lognormed_reads
condition_columns = ['tkov3_plasmid', 'dmso_1', 'dmso_2',
                     'dmso_3']
lognorm_df = tkov3_reads.copy()
lognorm_df[condition_columns] = lognorm_df[condition_columns].apply(lognorm)
condition_columns.remove("tkov3_plasmid")
lognorm_df
```

```python
#remove underrepresented sgRNAs
filtered_lognorms = lognorm_df.copy()
z_low = -3  # minimum z-score
# Z-score the pDNA column
filtered_lognorms["pDNA_zscore"] = (filtered_lognorms['tkov3_plasmid'] - filtered_lognorms['tkov3_plasmid'].mean())/filtered_lognorms['tkov3_plasmid'].std()
# Filter by z-score
filtered_lognorms = filtered_lognorms[filtered_lognorms["pDNA_zscore"] > z_low].reset_index(drop=True)
# Drop z-scored columns
filtered_lognorms = filtered_lognorms.drop("pDNA_zscore", axis=1)
filtered_lognorms
```

Calculate log fold change relative to pDNA (the later time point is after 14 population doublings, n days not reported) 

```python
lfc_df = filtered_lognorms.copy()
for col in condition_columns:
    lfc_df[col] = filtered_lognorms[col] - filtered_lognorms["tkov3_plasmid"]
lfc_df=lfc_df.drop("tkov3_plasmid",axis=1)
lfc_df
```

```python
#check replicate correlation
lfc_correlations = lfc_df.corr(numeric_only=True)
matrix=np.triu(lfc_correlations)
sns.clustermap(lfc_correlations, cmap='RdBu_r',
               vmin=-1, vmax=1, annot=True,row_cluster=False,col_cluster=False,mask=matrix)
```

```python
#verifies that this is a coincidence that they all round to 7.6 -- seemed suspicious
lfc_correlations
```

Since there is adequate replicate correlation for all constructs, we can average LFC for one value/guide

```python
lfc_df["sgRNA_lfc"]=lfc_df.mean(axis=1,numeric_only=True)
lfc_df=lfc_df.drop(condition_columns,axis=1)
lfc_df
```

The rows at the bottom are the  controls luciferase, EGFP, and LacZ, which should have no effect on viablity 


based on the [cited paper](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC6862721/) for analysis (which uses AAVS1 instead of luciferase/lacz/egfp), luciferase/lacz/egfp act as the controls to represent no effect **since they target sites that are not in the human genome**. 

It is important to note that, unlike intergenic controls, cells with these guides do not receive a double-stranded break. Thus, cells with targeting guides will inherently have lower viability from even a single-double stranded break. There should therefore be a stricter definition of promiscuous guides when using non-targeting controls rather than intergenics. 

```python
#identify the controls from the updated mappings provided by the LIMS
controls=lfc_df[lfc_df["Notes"].str.count("NO_SITE")>0].reset_index(drop=True).copy()
controls
```

```python
#verifying that controls are normally distributed with D'Agostino-Pearson K2 test
stat, p = normaltest(controls["sgRNA_lfc"])
print('All Control Statistics=%.3f, p=%.3f' % (stat, p), "length=", len(controls))
```

```python
controls.hist(grid=False)
```

The set of nontargeting controls are normally distributed

```python
#get average lfc across guides for each gene
lfc_df["gene_average_lfc"]= lfc_df.groupby('On-target Gene Symbols')["sgRNA_lfc"].transform('mean')


#subset data for nonessentials 
original_noness=lfc_df[lfc_df["noness_targeting"]].copy().reset_index(drop=True)

original_noness
```

Observing the mean log fold change across guides targeting a gene to see if any of the nonessential genes deplete (could be attributed to essential function or high copy number-- either way, this depletion does not reflect off-target activity)


Using [essential targeting genes](https://depmap.org/portal/data_page/?tab=allData&releasename=DepMap+Public+24Q2&filename=AchillesCommonEssentialControls.csv) (overlap between Biomen 2014 and Hart 2015) to identify the log fold change associated with perturbations that have an antiproliferative effect and recognize nonessential genes that should be reclassified 

```python
essentials=lfc_df[lfc_df["On-target Gene Symbols"].isin(essential_gene_list)].reset_index(drop=True).copy()
essentials_median=essentials["gene_average_lfc"].median()
```

```python
depleted_noness= original_noness[original_noness["gene_average_lfc"]<essentials_median]
noness_depleted_genes=depleted_noness["On-target Gene Symbols"].tolist()
depleted_noness
```

```python
#make list of nonessentials with depleted genes removed
noness= original_noness[-original_noness["On-target Gene Symbols"].isin(noness_depleted_genes)]
noness
```

As with the Rule Set 3 Validation set, we aim to identify the nonessential-targeting guides that have off-target activity. We expect that nonessential-targeting guides that cut at off-target sites will sustain a decrease in viability due to a toxic number of double-stranded breaks. For the rule set 3 validation set, we used the intergenic controls as a reference to identify guides that generate more than one double-stranded break, since intergenic controls reflect the viability effect of a single double-stranded break. In this dataset, there are no intergenic controls to reference. 

The spread of essential-targeting guides alone should not determine the promiscuous nonessential targeting guides, since **the spread of essential-targeting guides is determined by the design of the guide library**. 

Additionally, the spread of non-targeting guides alone should not determine the promiscuous nonessential targeting guides, since **nontargeting guides sustain 0 double stranded breaks and thus inherently have improved viability relative to nonessential-targeting guides**. We want to identify the guides that are likely to have sustained *more than* one double stranded break, as these are the ones involved in off-target activity.

In absence of intergenic controls, the most reliable manner by which to identify the promiscuous nonessential-targeting guides is to identify those that are ...
1. Just as or more depleted than the middle 68% of essential-targeting guides (= lower log fold change than 1 standard deviations above the mean of essential-targeting guides). For a guide to be promiscuous, it should have depletion levels comparable to the essential-targeting guides. 
2. More depleted than the middle 95% of nonessential-targeting guides (=2+ standard deviations below the nonessential-targeting guides mean). This should single out the "trail" of promiscuous guides that behave differently than the rest of nonessential-targeting guides

```python
essentials_mean=essentials["sgRNA_lfc"].mean(axis=0)
essentials_sd=essentials["sgRNA_lfc"].std(axis=0,numeric_only=True)
essentials_upper_limit= essentials_mean+(1*essentials_sd)
print("1 standard deviation above mean of essentials =",essentials_upper_limit)

noness_mean=noness["sgRNA_lfc"].mean(axis=0)
noness_sd=noness["sgRNA_lfc"].std(axis=0,numeric_only=True)
noness_lower_limit= noness_mean-(2*noness_sd)
print("2 standard deviations below mean of nonessentials =",noness_lower_limit)

noness_promiscuous_cutoff= min(essentials_upper_limit,noness_lower_limit)
print("lowest of all cutoffs to define promiscuous nonessentials",noness_promiscuous_cutoff)
```

```python
noness["promiscuous"]=noness["sgRNA_lfc"]<noness_promiscuous_cutoff
noness[noness["promiscuous"]==True]["sgRNA_lfc"].hist(grid=False,bins=100,label="promiscuous",color="tomato",edgecolor="tomato")
noness[noness["promiscuous"]==False]["sgRNA_lfc"].hist(grid=False,bins=100,label="remainder of guides",color="cornflowerblue",edgecolor="cornflowerblue")
plt.xlabel("Log fold change",fontsize=16)
plt.ylabel("# sgRNAs",fontsize=16)
plt.title("Nonessential-targeting sgRNAs, TKOv3",fontsize=16)
plt.legend(fontsize=16)
```

While there are some guides that seems to be more depleted than the rest of the group that are not labeled as promiscuous, the cutoff employed here gives us greater confidence that those labeled as promiscuous are actually reflecting off-target activity and not depleted by random chance.

Likewise, there should also be a data-driven approach to define non-promiscuous guides. We want to implement a strict threshold so that we can ensure that guides we are classifying as nonpromiscuous are truly unlikely to be experiencing a toxic number of double-stranded breaks. To be conservative, we require that a guide's log fold change is within the middle 68% of non-targeting control log fold changes in order to be defined as non-promiscuous. While this may be excluding some guides that are merely reflecting viabilty loss due to one double stranded break, this scenario is preferable to misclassifying a promiscuous guide as nonpromiscuous. Additionally, nonpromiscuous guides outnumber promiscuous guides (especially since the TKOv3 library was designed with off-target activity avoidance in mind), so losing nonpromiscuous guides minimally impacts the performance of the classifier. 

```python
nontargeting_mean=controls["sgRNA_lfc"].mean(axis=0)
nontargeting_sd=controls["sgRNA_lfc"].std(axis=0,numeric_only=True)
nontargeting_lower_limit= nontargeting_mean-(1*nontargeting_sd)
print("1 standard deviation below mean of nontargeting controls =",nontargeting_lower_limit)
noness_nonpromiscuous_cutoff= nontargeting_lower_limit
```

Guides that lie between the promiscuous and non-promiscuous threshold should be excluded from this analysis since it is uncertain if these guides have off-target activity. This gap enables more conservative estimates of which guides are promiscuous or not promiscuous

```python
noness_filt=noness[(noness["sgRNA_lfc"]<noness_promiscuous_cutoff) | (noness["sgRNA_lfc"]>noness_nonpromiscuous_cutoff)].reset_index(drop=True).copy()

noness_filt["promiscuous"]=noness_filt["sgRNA_lfc"]<noness_promiscuous_cutoff
noness_filt[noness_filt["promiscuous"]==True]["sgRNA_lfc"].hist(grid=False,bins=100,label="promiscuous",color="tomato",edgecolor="tomato")
noness_filt[noness_filt["promiscuous"]==False]["sgRNA_lfc"].hist(grid=False,bins=100,label="nonpromiscuous",color="cornflowerblue",edgecolor="cornflowerblue")
plt.xlabel("Log fold change",fontsize=16)
plt.ylabel("# sgRNAs",fontsize=16)
plt.title("Nonessential-targeting sgRNAs, TKOv3",fontsize=16)
plt.legend(fontsize=16)
plt.savefig("../Figures/tkov3_nonessentials_promiscuous.pdf", dpi=300, bbox_inches = 'tight')
```
