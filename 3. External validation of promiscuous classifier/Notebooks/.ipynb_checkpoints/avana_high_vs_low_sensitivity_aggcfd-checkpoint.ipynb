````python
import pandas as pd
import numpy as np
from sklearn.metrics import f1_score
import gzip
import matplotlib.pyplot as plt
import seaborn as sns
import gpplot as gpp
gpp.set_aesthetics(context = 'paper')
````

External validation of Aggregate CFD Score on cell lines stratified to exemplify the performance of this metric on cell lines with variable tolerance of double stranded breaks.

Load the nonessential-targeting guides cleaned in the prior jupyter notebook. These guides come from the avana library, and the screening data utilizes the median read counts across cell lines. 

# High sensitivity cell lines (TP53 WT)

````python
screening_data_high= pd.read_csv("../Data/avana_noness_high_sensitivity_screenresults.csv")
screening_data_high
````

Avana nonessential-targeting guides cleaned are copied into library mode of CRISPick to report all predicted off-target sites. This enables calculation of aggregate CFD of these guides. 

retrieving avana nonessentials scored for off-target activity from server: 

```scp ldrepano@login.broadinstitute.org:/broad/hptmp/gpp/sgrna-miner-extras/dev/sgrna-miner/59/37/5937db4e-b182-41ae-8b31-3bc9354ec95d/avana_noness_verbose-sgrna-designs.offtargetdisco.txt .```

compress file

``` gzip avana_noness_verbose-sgrna-designs.offtargetdisco.txt```

````python
with gzip.open('../Data/avana_noness_verbose-sgrna-designs.offtargetdisco.txt.gz') as f:
    crispick_verbose = (pd.read_table(f,index_col=False,header=None,
                                      names=["Target","Target Sequence","Context Sequence","Match Tier", "CFD Score","PAM Sequence", "# of mismatches","identifier"],
                                      usecols=["Target Sequence", "CFD Score", "# of mismatches"],
                                      on_bad_lines="warn"))
#get rid of NA pick orders (i.e. MAX OTS)
crispick_verbose=crispick_verbose[pd.to_numeric(crispick_verbose["# of mismatches"], errors='coerce').notnull()].reset_index(drop=True).copy()
crispick_verbose["# of mismatches"]=crispick_verbose["# of mismatches"].astype(np.int64)
crispick_verbose["CFD Score"]=crispick_verbose["CFD Score"].astype(np.float64)

crispick_verbose
````

````python
crispick_verbose["CFD_upto1mm"]= np.where(crispick_verbose["# of mismatches"]<=1, crispick_verbose["CFD Score"],0)


crispick_agg = (crispick_verbose.groupby("Target Sequence")
            .agg(aggcfd= ("CFD_upto1mm","sum"))
            .reset_index())
crispick_agg
````

````python
#merge with actual data
high_sensitivity=crispick_agg.merge(screening_data_high[["Target Sequence","promiscuous"]],on="Target Sequence")
high_sensitivity
````

````python
metric=high_sensitivity["aggcfd"].astype(np.float64)
f1_high=[]
cutoffs=[i*0.1 for i in range(300)]
for cutoff in cutoffs:
    metric_binary= metric > cutoff
    performance= f1_score(high_sensitivity["promiscuous"], metric_binary ,average="macro")
    f1_high.append(performance)
f1_high_best=max(f1_high) 
bestcutoff=cutoffs[f1_high.index(f1_high_best)]
high_sensitivity["Agg CFD Predicted"]=high_sensitivity["aggcfd"]>bestcutoff
print("f1_high= ", f1_score(high_sensitivity["promiscuous"], high_sensitivity["Agg CFD Predicted"] ,average="macro"), " at Agg CFD cutoff ",bestcutoff)
````

````python
pd.DataFrame(high_sensitivity[["Agg CFD Predicted","promiscuous"]].value_counts()).sort_values(by="Agg CFD Predicted")
````

````python
#true negative rate: TN/TN+FP
print("TNR:",1304/(1304+45))
#true positive rate: TP/TP+FN
print("TPR:",63/(63+52))
````

# Low sensitivity

````python
screening_data_low= pd.read_csv("../Data/avana_noness_low_sensitivity_screenresults.csv")
screening_data_low
````

````python
#merge with actual data
low_sensitivity=crispick_agg.merge(screening_data_low[["Target Sequence","promiscuous"]],on="Target Sequence")
low_sensitivity
````

````python
metric=low_sensitivity["aggcfd"].astype(np.float64)
f1_low=[]
for cutoff in cutoffs:
    metric_binary= metric > cutoff
    performance= f1_score(low_sensitivity["promiscuous"], metric_binary ,average="macro")
    f1_low.append(performance)
f1_low_best=max(f1_low) 
bestcutoff=cutoffs[f1_low.index(f1_low_best)]
low_sensitivity["Agg CFD Predicted"]=low_sensitivity["aggcfd"]>bestcutoff
print("f1_low= ", f1_score(low_sensitivity["promiscuous"], low_sensitivity["Agg CFD Predicted"] ,average="macro"), " at Agg CFD cutoff ",bestcutoff)
````

````python
pd.DataFrame(low_sensitivity[["Agg CFD Predicted","promiscuous"]].value_counts()).sort_values(by="Agg CFD Predicted")
````

````python
#true negative rate: TN/TN+FP
print("TNR:",1493/(1493+51))
#true positive rate: TP/TP+FN
print("TPR:",60/(60+47))
````

# Compare

````python
plt.plot(cutoffs,f1_low,label="TP53 Mutant Cell Lines (n=623)",linewidth=2,c="tomato")
plt.plot(cutoffs,f1_high,label="TP53 WT Cell Lines (n=384)",linewidth=2,c="cornflowerblue")

plt.xlim(0,12)
plt.ylim(0,1.1)
plt.ylabel("F1 Score",fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.axvline(x=4.8,c="black",linestyle="--",label="Aggregate CFD threshold")

plt.xlabel("CFD sum threshold to classify guide as promiscuous",fontsize=16)
plt.title("Aggregate CFD performance, Avana Data (DepMap)",fontsize=16)
plt.legend(fontsize=16,bbox_to_anchor=(1,1))
gpp.savefig("../Figures/tp53mut_vs_wt_aggcfd.pdf", dpi=600, bbox_inches = 'tight')
````
