```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import gpplot as gpp
from scipy.stats import normaltest
gpp.set_aesthetics(context = 'paper')
import gzip

```

Processing Sanger Guide Sequences (posted as 19nt) to incorporate in comparison of library drift

```python
sanger_library= pd.read_excel('../Data/Sanger/sanger_library.xlsx',sheet_name="Human v1")
sanger_library
```

Matching to genome-wide library guides to get full 20 nt

```python
with gzip.open('../Data/Sanger/sgRNA_design_9606_GRCh38_SpyoCas9_CRISPRko_RS3seq-Chen2013+RS3target_Ensembl_20240730.txt.gz') as f:
    full_genome_download = (pd.read_table(f))
full_genome_download
```

```python
full_genome_download["19nt guide"]=full_genome_download["sgRNA Sequence"].str[1:]
sanger_guides_20nt= full_genome_download[full_genome_download["19nt guide"].isin(sanger_library["Guide_sequence"].tolist())].reset_index(drop=True)[["sgRNA Sequence"]]
sanger_guides_20nt=sanger_guides_20nt.drop_duplicates(keep="first").reset_index(drop=True)
sanger_guides_20nt
```
