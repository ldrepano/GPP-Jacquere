```python
import pandas as pd
import numpy as np
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt
import seaborn as sns
import gpplot as gpp
gpp.set_aesthetics(context = 'paper')
```

External validation of Guidescan specificity score on cell lines stratified to exemplify the performance of the Specificity Score metric on cell lines with variable tolerance of double stranded breaks.

Load the nonessential-targeting guides cleaned in the prior jupyter notebook. These guides come from the avana library, and the screening data utilizes the median read counts across cell lines. 

# High sensitivity cell lines (TP53 WT)

```python
screening_data_high= pd.read_csv("../Data/avana_noness_high_sensitivity_screenresults.csv")
screening_data_high
```

```python
screening_data_high["promiscuous"].value_counts()
```

### Get guidescan scores

Guidescan defines a promiscuous guide as having:

1. any OTS with 0 mismatches to target site
2. more than 1 OTS with 1 mismatch to target site 
3. CFD sum of OTS with up to 3 mismatches of >=4 (this is equivalent to specificity score 0.2) 

and includes no OTS with an NGG PAM in this search. Mismatch # includes any mismatch in entire 20mer.

Upon search using the Guidescan tool, guides with criteria 1 or 2 are not evaluated. Since these would be filtered out in screen design, these are treated as having a specificity score of 0 for this analysis.

```python
#ran guidescan on all avana guides (essential included)
guidescan=pd.read_csv("../Data/guidescan_avana_results.csv")

#specificity=0 if it violates the 0 or 1mm rule (therefore is 'not in database')
guidescan["guidescan_specificity"]=np.where(guidescan["Chromosome"].str.contains("chr"), guidescan["Specificity"],0)
guidescan["Target Sequence"]=guidescan["gRNA"].str[:20]

guidescan
```

```python
#merge with actual data
high_sensitivity=guidescan.merge(screening_data_high[["Target Sequence","promiscuous"]],on="Target Sequence")
high_sensitivity
```

```python
metric=high_sensitivity["guidescan_specificity"].astype(np.float64)
f1_high=[]
cutoffs=[i*0.01 for i in range(100)]
for cutoff in cutoffs:
    metric_binary= metric < cutoff
    performance= f1_score(high_sensitivity["promiscuous"], metric_binary ,average="macro")
    f1_high.append(performance)
f1_high_best=max(f1_high) 
bestcutoff=cutoffs[f1_high.index(f1_high_best)]
high_sensitivity["Guidescan Predicted"]=high_sensitivity["guidescan_specificity"]<bestcutoff
print("f1_high= ", f1_score(high_sensitivity["promiscuous"], high_sensitivity["Guidescan Predicted"] ,average="macro"), " at Guidescan cutoff ",bestcutoff)
```

```python
high_sensitivity["Guidescan Predicted at 0.2 cutoff"]=high_sensitivity["guidescan_specificity"]<0.2
pd.DataFrame(high_sensitivity[["Guidescan Predicted at 0.2 cutoff","promiscuous"]].value_counts()).sort_values(by="Guidescan Predicted at 0.2 cutoff")
```

```python
#true negative rate: TN/TN+FP
print("TNR:",546/(546+756))
#true positive rate: TP/TP+FN
print("TPR:",143/(143+3))
```

# Compare

```python
#convert specificity score (1/ 1+ CFD sum) to the corresponding CFD sum 
cutoffs_cfdsum=[(-1 + 1/i) for i in cutoffs[1:] ]

plt.plot(cutoffs_cfdsum,f1_low[1:],label="TP53 Mutant Cell Lines (n=623)",linewidth=2,c="tomato")
plt.plot(cutoffs_cfdsum,f1_high[1:],label="TP53 WT Cell Lines (n=384)",linewidth=2,c="cornflowerblue")

plt.ylim(0,1.1)
plt.xlim(0,20)

plt.ylabel("F1 Score",fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.axvline(x=4,c="black",linestyle="--",label="Guidescan2 threshold")
plt.xlabel("CFD sum threshold to classify guide as promiscuous",fontsize=16)
plt.title("Guidescan Specificity Score performance, Avana Data (DepMap)",fontsize=16)
plt.legend(fontsize=16,bbox_to_anchor=(1,1))
gpp.savefig("../Figures/tp53mut_vs_wt_guidescan.pdf", dpi=600, bbox_inches = 'tight')
```
